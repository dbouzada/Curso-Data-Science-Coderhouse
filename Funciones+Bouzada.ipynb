```python
#Ejercicio 1: Mati, para cambiar el cálculo del factorial habría que modificar el número entre parentesis del print.
#Si, recordá que la función es algo abstracto=> Vos podrías reemplazar "n" por cualquier nnumero y tendría que andar
#Éstá bien esto, faltaría agregar que si n es < 0 te diga que no se puede calcular

def fac(n):
  if n <0: 
    return "No se puede calcular"
   #acá un condicional de si n es menor a 0 devuelva "No se puede calcular"
  else:
    numero_ = 1
    for i in range(1, n+1):
      numero_ *= i
      
  return numero_
    
print (fac (-8))
print (fac (0))
print (fac (5))
```

```python
def unaFuncionConDosParametros( parametro1, parametro2):
  print(f'El parametro1 es {parametro1}') #Verás que no sabemos qué es parametro1 pero queremos que lo muestre acá
  print(f'El parametro2 es {parametro2}')
  resultado=parametro1+parametro2 #La suma de estos parámetros es el resultado...pero no sabemos cuáles son
  print(f'El resultado es {resultado}')
  return resultado

unaFuncionConDosParametros(2,2) #Acá la llamamos y le pasamos un valor a cada parametro
print("###")
unaFuncionConDosParametros(20,200) #La utilidad de las funciones es que creas una maquina que recibe parametros y devuelve resultados
print("###")
unaFuncionConDosParametros(100,-200)
```

```python
#Ejercicio 2
#
#
#
 
def suma_intervalo(num1,num2):
  if num2<num1:
    return "El primer numero debe ser menor que el segundo, revise su parametro"
  else:
    suma = 0
    for i in range(num1,num2+1): #1 tiene que ser num1 y 101 num2 +1
      suma += i
  return suma
  
print(suma_intervalo (1,100)) #Te falta pasarle los parámetros ej; suma_intervalo(1,100)
#Excelente! Ves ahora cómo usar los parámetros dentro del contexto de la función?
#Lo unico que agregaría ahora seria una validacion como en la de arriba para que si num2 es menor a num1 tire el error
print(suma_intervalo(100,1))
print(suma_intervalo (1,50))
```
