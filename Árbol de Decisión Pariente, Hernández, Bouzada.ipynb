# Segundo Ensayo
### Aquí intentamos con cuatro estaciones y nos tomó todo bien

```python
df_estaciones = df_bicis[df_bicis.id_estacion_origen.isin([14 , 33 , 9, 29]) & #Los nombres son: Facultad de Medicina" , "Pacífico", "Parque Las Heras", "Plaza Italia"
               df_bicis.id_estacion_destino.isin([14 , 33, 9, 29])] # Los nombres son: "Facultad de Medicina" , "Pacífico", "Parque Las Heras", "Parque Centenario"
df_estaciones.head()
```

```python
df_estaciones3 = df_estaciones.iloc[:, [0,2,5,9]]
df_estaciones3.head()
```

```python
bicis_dummies2 = pd.get_dummies(df_estaciones3, columns = ["id_usuario", "id_estacion_origen", 'id_estacion_destino'])
bicis_dummies2
```

```python
X = bicis_dummies2
y = df_estaciones3.tiempo_recorrido
```

```python
from sklearn.model_selection import train_test_split #Importamos la libreria para separar en train y test

#Me quedo con 30% para test y 70% para train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) 
```

```python
from sklearn.tree import DecisionTreeClassifier #Importamos el modelo
tree = DecisionTreeClassifier(max_depth=20, random_state = 42) #Creamos el modelo
```

```python
#Entrenamos el modelo
tree.fit(X_train,y_train) #Entrenamos el modelo
```

```python
y_train_pred = tree.predict(X_train) #Prediccion en Train
y_test_pred = tree.predict(X_test) #Prediccion en Test
```

```python
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de entrenamiento:', train_accuracy)
print('% de aciertos sobre el set de evaluación:',test_accuracy)
```
