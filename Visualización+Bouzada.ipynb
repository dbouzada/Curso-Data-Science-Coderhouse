#Ejercicio 4

```python
empleados.DateofHire
```

```python
empleados ["DateofHire"] = empleados["DateofHire"].apply(lambda x: pd.to_datetime(x, format= ("%m/%d/%Y")))
empleados.head()
```

```python
empleados.dtypes
```

```python
fecha_contratacion = empleados.groupby (empleados["DateofHire"].dt.year) ["EmpID"].count()
fecha_contratacion
```

```python
ax = sns.lineplot (data=fecha_contratacion)
ax.set(xlabel='Año de Contratación', ylabel='Cantidad de Empleados', title='Serie de tiempo por fecha de contratación')
```
